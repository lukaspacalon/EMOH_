```python
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import sent_tokenize
from colorama import Fore

def check_sentiment(text):
    sia = SentimentIntensityAnalyzer()
    sentences = sent_tokenize(text)
    
    for sentence in sentences:
        sentiment_scores = sia.polarity_scores(sentence)
        if sentiment_scores['compound'] >= 0:
            print(Fore.BLUE + sentence)
        else:
            print(Fore.RED + sentence)

# Exemple d'utilisation
texte = """
savait, et l'ayant psalmodié douze fois, il fit une
afin
douze
Paul. [7J Mais celuifois,
d'éprouver
prière
ci de son côté s'unissait avec ardeur à la prière, car, à ce
que je pense, il eût préféré paître des scorpions que de
vivre avec une femme adultère. Cependant, après les
douze prières, le soir étant avancé, ils s'assirent pour
ayant mangé l'un des biscuits,
manger. Or Antoine,
ne toucha pas à un autre. Mais le vieillard,
qui mangeait plus lentement, en était encore à son petit biscuit.
Antoine attendait qu'il eût fini, et il lui dit : « Mange,
» Paul lui dit :
petit père, encore un autre biscuit.
H Dans le cas où tu en manges, moi aussi ; mais toi ne
pas, je ne mange pas. » Antoine lui dit:
mangeant
qu'il

« Cela me suffit, car je suis moine. » [8] Paul lui dit :
« 11 me suffit également, car moi aussi je veux devenir
moine. » Il s'éveille de nouveau et fait douze prières et
psalmodie douze psaumes. 11 dort un peu du premier
au
sommeil, et de nouveau s'éveille pour psalmodier
milieu de la nuit jusqu'au jour. Alors, comme il voyait
que le vieillard l'avait suivi avec ardeur dans son régime,
il lui dit: « Si tu peux ainsi jour par jour,reste avec-moi.»
Paul lui dit: « En vérité, si parfois il arrive qu'il y ait
ce
quelque chose de plus, je ne sais pas ; car autrement,
» Antoine lui dit
que j'ai vu, je le fais bien dexlrement.
le (jour) après : « Voici que tu es devenu moine ».
"""
check_sentiment(texte)
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

def detecter_emotion(texte):
    sia = SentimentIntensityAnalyzer()
    phrases = nltk.sent_tokenize(texte)  # Diviser le texte en phrases
    
    for phrase in phrases:
        polarite = sia.polarity_scores(phrase)
        
        if polarite['compound'] >= 0.5:
            emotion = "\033[93m"  # Affichage en jaune pour la joie
        elif polarite['compound'] <= -0.5:
            emotion = "\033[94m"  # Affichage en bleu pour la tristesse
        elif polarite['compound'] > -0.5 and polarite['compound'] < 0.5:
            if polarite['neg'] > polarite['pos']:
                emotion = "\033[91m"  # Affichage en rouge pour la colère
            elif polarite['neg'] > polarite['neu']:
                emotion = "\033[95m"  # Affichage en violet pour la peur
            else:
                emotion = "\033[92m"  # Affichage en vert pour le neutre
        
        print(f"{emotion}{phrase}\033[0m")


# Exemple d'utilisation
texte = """

savait, et l'ayant psalmodié douze fois, il fit une
afin
douze
Paul. [7J Mais celuifois,
d'éprouver
prière
ci de son côté s'unissait avec ardeur à la prière, car, à ce
que je pense, il eût préféré paître des scorpions que de
vivre avec une femme adultère. Cependant, après les
douze prières, le soir étant avancé, ils s'assirent pour
ayant mangé l'un des biscuits,
manger. Or Antoine,
ne toucha pas à un autre. Mais le vieillard,
qui mangeait plus lentement, en était encore à son petit biscuit.
Antoine attendait qu'il eût fini, et il lui dit : « Mange,
» Paul lui dit :
petit père, encore un autre biscuit.
H Dans le cas où tu en manges, moi aussi ; mais toi ne
pas, je ne mange pas. » Antoine lui dit:
mangeant
qu'il

« Cela me suffit, car je suis moine. » [8] Paul lui dit :
« 11 me suffit également, car moi aussi je veux devenir
moine. » Il s'éveille de nouveau et fait douze prières et
psalmodie douze psaumes. 11 dort un peu du premier
au
sommeil, et de nouveau s'éveille pour psalmodier
milieu de la nuit jusqu'au jour. Alors, comme il voyait
que le vieillard l'avait suivi avec ardeur dans son régime,
il lui dit: « Si tu peux ainsi jour par jour,reste avec-moi.»
Paul lui dit: « En vérité, si parfois il arrive qu'il y ait
ce
quelque chose de plus, je ne sais pas ; car autrement,
» Antoine lui dit
que j'ai vu, je le fais bien dexlrement.
le (jour) après : « Voici que tu es devenu moine ».
"""

detecter_emotion(texte)
```

```python

from text2emotion import get_emotion
from termcolor import colored

def afficher_texte_emotion(texte):
    phrases = texte.split('. ')
    
    for phrase in phrases:
        emotion = get_emotion(phrase)
        couleur = obtenir_couleur_emotion(emotion)
        texte_colore = colored(phrase, couleur)
        print(couleur)
        print(texte_colore)

def obtenir_couleur_emotion(emotion):
    if emotion == "Happy":
        return "green"
    elif emotion == "Angry":
        return "red"
    elif emotion == "Sad":
        return "blue"
    elif emotion == "Surprise":
        return "yellow"
    elif emotion == "Fear":
        return "magenta"
    else:
        return "white"

# Exemple d'utilisation
texte = """

savait, et l'ayant psalmodié douze fois, il fit une
afin
douze
Paul. [7J Mais celuifois,
d'éprouver
prière
ci de son côté s'unissait avec ardeur à la prière, car, à ce
que je pense, il eût préféré paître des scorpions que de
vivre avec une femme adultère. Cependant, après les
douze prières, le soir étant avancé, ils s'assirent pour
ayant mangé l'un des biscuits,
manger. Or Antoine,
ne toucha pas à un autre. Mais le vieillard,
qui mangeait plus lentement, en était encore à son petit biscuit.
Antoine attendait qu'il eût fini, et il lui dit : « Mange,
» Paul lui dit :
petit père, encore un autre biscuit.
H Dans le cas où tu en manges, moi aussi ; mais toi ne
pas, je ne mange pas. » Antoine lui dit:
mangeant
qu'il

« Cela me suffit, car je suis moine. » [8] Paul lui dit :
« 11 me suffit également, car moi aussi je veux devenir
moine. » Il s'éveille de nouveau et fait douze prières et
psalmodie douze psaumes. 11 dort un peu du premier
au
sommeil, et de nouveau s'éveille pour psalmodier
milieu de la nuit jusqu'au jour. Alors, comme il voyait
que le vieillard l'avait suivi avec ardeur dans son régime,
il lui dit: « Si tu peux ainsi jour par jour,reste avec-moi.»
Paul lui dit: « En vérité, si parfois il arrive qu'il y ait
ce
quelque chose de plus, je ne sais pas ; car autrement,
» Antoine lui dit
que j'ai vu, je le fais bien dexlrement.
le (jour) après : « Voici que tu es devenu moine ».

"""
afficher_texte_emotion(texte)
```
